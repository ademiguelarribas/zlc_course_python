## 3. Running Python: Interactive vs Script

There are two main ways of running Python code:

### 1. Interactive (what we are doing now)
- Jupyter Notebook / Google Colab
- Python REPL
- VS Code interactive window

You type a command ‚Üí Python runs it immediately ‚Üí you see the result.

### 2. Script
A **script** is a file ending in `.py`, for example:

```bash
python myscript.py
```

We will mostly use **notebooks** in this course, but it is important to know that scripts exist and are common in real projects.

## 4. How Python Sees a Program

- Python executes instructions in **sequence**.
- Blank lines are ignored (you can use them to separate ideas).
- **Indentation (spaces at the start of a line) is syntax.**
- Indentation groups related lines, for example under an `if` or a loop.

We will explore `if` and loops in later sessions, but here is a small taste:


````python
temperature = 5

if temperature > 0:
    print("Above freezing!")
````

## 5. Comments

A **comment** is text that Python **ignores**.

We use comments to explain what the code does, to leave notes, or to temporarily disable code.

- A comment starts with `#`.
- Everything after `#` on that line is ignored by Python.


````python
# This line prints a message
print("Hello again!")

# print("This line will not run because it is commented out")
````

## 6. Variables: Names for Values

A **variable** is a name that refers to a value.

We create variables using the **assignment operator** `=`:

```python
a = 5
b = 2
```

Then we can use these names in expressions.


````python
a = 5
b = 2

print("a =", a)
print("b =", b)
print("Sum:", a + b)
````

üí° **Tip:** use clear, lowercase names:

- ‚úÖ `price`, `total_kg`, `num_trucks`
- ‚ùå `X1`, `a1b2c3`

Good names make your code easier to read (for others and for future you).


---
## 7. What Happens in Memory (Intuition)

When you write:

```python
x = 10
```

Python:

1. Creates an object in memory with value `10`.
2. Makes the variable name `x` **point** (refer) to that object.

You can think of variables as **labels attached to values**.

Don't worry about the details yet ‚Äî we will revisit this more formally in Session 1.
